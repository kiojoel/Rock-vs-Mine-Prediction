The goal of this project is to develop a machine learning model that can accurately classify objects detected by sonar signals as either rocks or mines. The model will be trained and evaluated using the SONAR dataset, which contains sonar signal data.


Importing Dependecies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and Processing

```python
# Loading dataset to a pandas DataFrame
dataset = pd.read_csv('/content/sonar data.csv', header=None)
```

```python
dataset.head()
```

```python
# size of the dataset (rows, columns)
dataset.shape
```

```python
# Dataset Statistics
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset[60].value_counts()
```

Mine(M) ---> 111

Rock(R) ---> 97



```python
# grouping the data according to the mean
dataset.groupby(60).mean()
```

```python
# Seperating data and label
X = dataset.drop(columns=60, axis=1)
Y = dataset[60]
```

```python
print(X)
```

```python
print(Y)
```

Spliting data into Training and Test data

```python
X_train, X_test, Y_train , Y_test = train_test_split(X,Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape , Y_test.shape )
```

We have 187 train data and 21 test data.


**Model Training** with Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy on training data
train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(train_prediction, Y_train)
```

```python
print('Accuracy of training data: ', training_data_accuracy)
```

```python
# accuracy on test data
test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(test_prediction, Y_test)
```

```python
print('Accuracy of test data: ', testing_data_accuracy)
```
